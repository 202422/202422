```python
from sklearn.model_selection import train_test_split, GridSearchCV
import pandas as pd
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
```

```python
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("/content/dataset_features_handled.csv")
```

```python
df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop("Delivery_Duration", axis = 1), df["Delivery_Duration"], test_size=0.2)
```

```python
X_test
```

```python
Y_train
```

```python
df.corr()["Delivery_Duration"].sort_values(ascending = False)
```

```python
important_feat = ["estimated_store_to_consumer_driving_duration", "subtotal", "total_outstanding_orders", "estimated_order_place_duration",
                  "max_item_price", "num_distinct_items", "created_at_hour", "creation_period_day" ]
```

```python
model_svr = SVR()

param_grid = {"kernel": ["rbf", "linear", "poly"],
              "C": [0.001, 0.01, 0.1, 1, 10, 100]
              }

grid = GridSearchCV(model_svr, cv = 5, param_grid=param_grid)
```

```python
model_svr_simple = SVR(kernel = "poly", C = 1)
model_svr_simple.fit(X_train[important_feat], Y_train)
```

```python
model_randomF_simple = RandomForestRegressor(random_state=3, n_estimators=300, n_jobs=-1, verbose = 2)
model_randomF_simple.fit(X_train[important_feat], Y_train)
```

```python
model_ridge = Ridge()
model_ridge.fit(X_train, Y_train)
```

```python
model_adaboost = AdaBoostRegressor(n_estimators=300)
model_adaboost.fit(X_test, Y_test)
```

```python
model_adaboost.score(X_test, Y_test)
```

```python
model_ridge.score(X_test, Y_test)
```

```python
model_randomF_simple2 = RandomForestRegressor(random_state=3)

# Hyperparameter grid
param_grid = {
    "n_estimators": [200, 300],         # number of trees
    "max_depth": [None, 10, 20, 30],         # tree depth
    "min_samples_split": [2, 5, 10],         # min samples to split a node
    "min_samples_leaf": [1, 2, 4],           # min samples at a leaf
    "max_features": ["sqrt", "log2", None],   # number of features to consider
    "verbose" : [2]
}

# GridSearchCV setup
grid_search = GridSearchCV(
    estimator=model_randomF_simple2,
    param_grid=param_grid,
    cv=4,              # 3-fold cross validation
    n_jobs=-1,
    verbose=2,
    scoring="r2"       # or "neg_mean_squared_error" depending on your goal
)

# Fit
grid_search.fit(X_train[important_feat], Y_train)
```

```python
model_svr_simple.score(X_test[important_feat], Y_test)
```

```python
model_randomF_simple.score(X_test[important_feat], Y_test)
```
